### **Outliers Values**

Terlihat pada kolom:

- No_of_Votes

  Dengan persentase sebesar 6.7%, merupakan kolom dengan jumlah outliers tertinggi.


- IMDB_Rating
  
  Dengan persentase sebesar 1.3%.

- Meta_score

  Dengan persentase sebesar 0.7% (di bawah 1%), merupakan kolom dengan jumlah outliers paling rendah.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Titik-titik di bawah ini menunjukkan adanya data dengan jumlah voting yang jauh lebih tinggi dibandingkan sebagian besar film lainnya.

Hal ini sesuai dengan hasil perhitungan sebelumnya, dimana kolom No_of_Votes memiliki persentase outliers sebesar 6.7%, yang merupakan yang tertinggi dibandingkan kolom lainnya.


```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['No_of_Votes'])
plt.title('Boxplot of No_of_Votes')
plt.ylabel('No_of_Votes')
plt.show()
```

## **Eksplorasi Data (EDA)**

### **Comparison/Perbandingan**

Comparison dalam analisis IMDb Movies Dataset digunakan untuk membandingkan nilai antar kategori agar terlihat mana yang lebih unggul. Pada grafik Top 10 Genre dengan Rata-rata IMDb Rating Tertinggi, terlihat bahwa genre Animation, Drama, War memiliki rata-rata rating lebih tinggi dibanding genre lain seperti Adventure, Mystery, Thriller. Hal ini menunjukkan bahwa setiap genre memiliki kualitas yang berbeda menurut penilaian penonton.
Analisis perbandingan seperti ini penting karena membantu mengidentifikasi kategori yang menonjol, baik dari sisi rating maupun votes. Dengan begitu, kita bisa mengetahui genre atau aktor mana yang lebih konsisten menghasilkan film berkualitas tinggi, sehingga dapat dijadikan acuan dalam pengambilan keputusan.


```python
rating_genre = df.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False)

top10_genre = rating_genre.head(10)

plt.figure(figsize=(10,6))
sns.barplot(y=top10_genre.index, x=top10_genre.values, palette='viridis')
plt.title('Top 10 Genre dengan Rata-rata IMDB Rating Tertinggi')
plt.xlabel('Rata-rata IMDB Rating')
plt.ylabel('Genre')
plt.show()
```

Composition pada analisis IMDb Movies Dataset digunakan untuk melihat proporsi atau kontribusi suatu kategori terhadap keseluruhan data. ### **Relationshio/Hubungan**

Relationship dalam analisis IMDb Movies Dataset digunakan untuk melihat hubungan antar variabel, apakah satu variabel memengaruhi atau berkaitan dengan variabel lain. Contohnya, pada heatmap Korelasi antara Jumlah Votes dan IMDb Rating, terlihat nilai korelasi sebesar 0.49. Angka ini menunjukkan adanya hubungan positif sedang antara jumlah votes dengan rating, semakin banyak penonton yang memberikan suara, cenderung rating film juga lebih tinggi, meskipun tidak selalu kuat.
Analisis relationship seperti ini penting karena membantu memahami keterkaitan antara popularitas dan kualitas. Dengan mengetahui adanya korelasi, industri perfilman bisa menilai bahwa film yang populer (banyak votes) umumnya juga dianggap berkualitas oleh penonton. Namun, karena korelasinya tidak terlalu tinggi, hal ini juga menegaskan bahwa popularitas tidak selalu identik dengan kualitas


```python
plt.figure(figsize=(8, 6))
sns.heatmap(
    data=df[['No_of_Votes', 'IMDB_Rating']].corr(),
    annot=True,
    cmap='viridis',
    fmt='.2f'
)
plt.title('Korelasi antara Jumlah Votes dan IMDb Rating')
plt.show()
```
