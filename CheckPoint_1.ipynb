## **Business Objective**

IMDb Movies Dataset

Data ini merupakan sebuah data yang berisi 1000 film dan acara TV terbaik berdasarkan rating IMDb, yang mencakup informasi seperti judul, genre, tahun rilis, durasi, rating, jumlah voting dari penonton, serta tokoh atau pemain film. Data ini mempresentasikan karya hiburan dengan tingkat popularitas dan kualitas yang tinggi menurut penilaian publik.

Permasalahan yang muncul adalah belum optimalnya pemanfaatan data dalam menganalisis pola kesukaan penonton, seperti genre yang paling disukai, durasi film yang cenderung mendapatkan rating tinggi, peran pemain film terhadap performa rating, serta karakteristik film yang memiliki tingkat popularitas besar. Banyak keputusan produksi masih didasarkan pada tren umum tanpa analisis data yang mendalam.

Dengan menganalisis data ini, diharapkan industri perfilman dapat memperoleh gambaran yang jelas mengenai preferensi penonton sehingga dapat digunakan sebagai bahan pertimbangan dalam membuat film, menentukan konsep cerita, memilih pemain yang sesuai, serta menyusun strategi produksi yang lebih repat dan berbasis data.
